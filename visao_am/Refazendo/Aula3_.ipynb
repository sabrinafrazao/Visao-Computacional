```python
img1 = cv2.imread(path + 'gato1.jpg' )
```

```python
img1_c = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
```

```python
def plot_img2(img1, title1, img2, title2, tam):
    figure(figsize = tam)
    plt.subplot(1, 2, 1), plt.imshow(img1), plt.title(title1)
    plt.xticks([]), plt.yticks([])
    plt.subplot(1, 2, 2), plt.imshow(img2), plt.title(title2)
    plt.xticks([]), plt.yticks([])
    plt.show()
```

```python
plot_img2(img1, "Sem conversão", img1_c, "Convertida", (12, 10))
```

```python
im_olho = img1_c[150:200, 140:190]
plt.imshow(im_olho)
```

```python
def plot_img4(img1, title1, img2, title2, img3, title3, img4, title4, tam, cmp=True):
    figure(figsize=tam)
    plt.subplot(1, 4, 1), plt.imshow(img1, cmap='gray'), plt.title(title1)
    plt.xticks([]), plt.yticks([])

    if cmp:
        plt.subplot(1, 4, 2), plt.imshow(img2, cmap='gray'), plt.title(title2)
    else: 
        plt.subplot(1, 4, 2), plt.imshow(img2), plt.title(title2)
        plt.xticks([]), plt.yticks([])
    if cmp:
        plt.subplot(1, 4, 3), plt.imshow(img3, cmap='gray'), plt.title(title3)
    else:
        plt.subplot(1, 4, 3), plt.imshow(img3), plt.title(title3)
        plt.xticks([]), plt.yticks([])
    if cmp:
        plt.subplot(1, 4, 4), plt.imshow(img4, cmap='gray'), plt.title(title4)
    else:
        plt.subplot(1, 4, 4), plt.imshow(img4), plt.title(title4)
        plt.xticks([]), plt.yticks([])

    plt.show()
        
```

```python
img1_c_b, img1_c_g, img1_c_r = cv2.split(img1_c)
```

```python
plot_img4(img1_c, "Original", img1_c_b, "B", img1_c_g, "G", img1_c_r, "R", (12, 10))
```
