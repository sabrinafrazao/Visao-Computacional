```python
import cv2
from PIL import Image, ImageOps
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from pylab import *
```

```python
k_sobel_x = np.array([[-1, 0,1],
                     [-2, 0, 2],
                     [-1, 0, 1]], dtype=np.float32)

k_sobel_y = np.array([[-1, -2, -1],
                     [0, 0, 0],
                     [1, 2, 1]], dtype=np.float32)

k_laplace = np.array([[0, 1, 0],
                     [1, -4, 1],
                     [0, 1, 0]], dtype=np.float32)

k_h = np.array([[-1, -1, -1],
                     [2, 2, 2],
                     [-1, -1, -1]], dtype=np.float32)

k_v = np.array([[-1, 2, -1],
                [-1, 2, -1],
                     [-1, 2, -1]], dtype=np.float32)

k_sharp = np.array([[0, -1, 0],
                     [-1, 5, -1],
                     [0, -1, 0]], dtype=np.float32)

k_avg_blur = np.ones((5,5), np.float32)/25

k_emboss = np.array([[2, -1, 0],
                     [-1, 1, 1],
                     [0, 1, 2]], dtype=np.float32)
```

```python
video2 = path + 'video2.mp4'
```

```python
def aplica_sobel(im_gray):
    sobel_x = cv2.filter2D(im_gray, -1, k_sobel_x)
    sobel_y = cv2.filter2D(im_gray, -1, k_sobel_y)
    return cv2.magnitude(np.float32(sobel_x), np.float32(sobel_y))

def aplica_laplace(im_gray):
    return cv2.filter2D(im_gray, -1, k_laplace)

def aplica_linha_h(im_gray):
    return cv2.filter2D(im_gray, -1, k_h)

def aplica_linha_v(im_gray):
    return cv2.filter2D(im_gray, -1, k_v)

def aplica_sharp(im_frame):
    return cv2.filter2D(im_frame, -1, k_sharp)
    
def aplica_blur(im_frame):
    return cv2.filter2D(im_frame, -1, k_avg_blur)

def aplica_emboss(im_frame):
    return cv2.filter2D(im_frame, -1, k_emboss)

def aplica_cor(frame, limiar_claro, limiar_escuro):
    cor_hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    mask = cv2.inRange(cor_hsv, limiar_claro, limiar_escuro)
    saida = cv2.bitwise_and(frame, frame, mask=mask)
    return saida
```

```python
def processa_frame(frame, modo):
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    if modo == 'grayscale':
        return gray
    elif modo == 'sobel':
        return aplica_sobel(gray)
    elif modo == 'laplace':
        return aplica_laplace(gray)
    elif modo == 'linhas_h':
        return aplica_linha_h(gray)
    elif modo == 'linhas_v':
        return aplica_linha_v(gray)
    elif modo == 'sharp':
        return aplica_sharp(frame)
    elif modo == 'blur':
        return aplica_blur(frame)
    elif modo == 'emboss':
        return aplica_emboss(frame)
    elif modo == 'azul':
        limiar_claro = np.array([90, 50, 50])
        limiar_escuro = np.array([128, 255, 255])
        return aplica_cor(frame, limiar_claro, limiar_escuro)
    elif modo == 'marrom':
        limiar_claro = np.array([10, 100, 20])
        limiar_escuro = np.array([20, 255, 200])
        return aplica_cor(frame, limiar_claro, limiar_escuro )
    else:
        return frame
```

```python
import time
def main():
    modo = 'color'
    cap = cv2.VideoCapture(video2)

    cv2.namedWindow('Quadro', cv2.WINDOW_NORMAL)
    cv2.resizeWindow('Quadro', 640, 600)

    while True:
        rval, frame = cap.read()
        if not rval:
            break
        quadro_real = processa_frame(frame, modo)

        cv2.imshow('Quadro', quadro_real)
        time.sleep(0.3)

        key =  cv2.waitKey(1) & 0xFF

        if key == ord('g'):
            modo = 'grayscale'
        elif key == ord('s'):
            modo = 'sobel'
        elif key == ord('l'):
            modo = 'laplace'
        elif key == ord('h'):
            modo = 'linhas_h'
        elif key == ord('v'):
            modo = 'linhas_v'
        elif key == ord('p'):
            modo = 'sharp'
        elif key == ord('b'):
            modo = 'blur'
        elif key == ord('e'):
            modo = 'emboss'
        elif key == ord('c'):
            modo = 'color'
        elif key == ord('a'):
            modo = 'azul'
        elif key == ord('m'):
            modo = 'marrom'
        elif key == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```
