```python
import cv2
from PIL import Image, ImageOps
import matplotlib.pyplot as plt
from pylab import *
import numpy as np
```

```python
def plot_im3_(im1, tit1, im2, tit2, im3, tit3, tam):
    figure(figsize = tam)
    plt.subplot(1, 3, 1), plt.imshow(im1), plt.title(tit1)
    plt.xticks([]), plt.yticks([])
    
    plt.subplot(1, 3, 2), plt.imshow(im2), plt.title(tit2)
    plt.xticks([]), plt.yticks([])

    plt.subplot(1, 3, 3), plt.imshow(im3), plt.title(tit3)
    plt.xticks([]), plt.yticks([])
    plt.show()
```

```python
LuzSemFleshO = cv2.imread(path + 'LuzSemFleshO.jpg' )
LuzSemf = cv2.imread(path + 'LuzSemf.jpg' )
```

```python
LuzSemFleshO_c = cv2.cvtColor(LuzSemFleshO, cv2.COLOR_BGR2RGB)
LuzSemf_c = cv2.cvtColor(LuzSemf, cv2.COLOR_BGR2RGB)
```

```python
dif_abs = cv2.absdiff(LuzSemFleshO_c, LuzSemf_c)
```

```python
plot_im3_(LuzSemFleshO_c, 'IMG1', LuzSemf_c, 'IMG2', dif_abs, 'Diferença', (12, 10))
```
