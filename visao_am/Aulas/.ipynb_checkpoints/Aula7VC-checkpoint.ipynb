```python
import numpy as np
import tensorflow.keras as tk
import keras
import keras.datasets as ds
from keras import layers
from datetime import datetime
```

```python
num_classes = 10
```

```python
input_shape = (28,28,1)
```

```python
mnist = tk.datasets.mnist
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(8, 6))
fig.subplots_adjust(left = 0, right = 1, bottom =0, top = 1, hspace = 0.05, wspace = 0.05)

for i in range(25):
    ax = fig.add_subplot(5, 5, i+1, xticks=[], yticks=[])
    ax.imshow(X_train[i], cmap=plt.cm.gray, interpolation='lanczos')
```

```python
X_train = np.expand_dims(X_train, -1)
X_test = np.expand_dims(X_test, -1)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
y_train = tk.utils.to_categorical(y_train, num_classes)
y_test = tk.utils.to_categorical(y_test, num_classes)
```

```python
y_train.shape, y_test.shape
```

```python
modelo = keras.Sequential(
    [
        keras.Input(shape=input_shape),
        layers.Conv2D(32, kernel_size = (3 ,3), activation="relu"),
        layers.MaxPooling2D(pool_size=(2,2)),
        layers.Conv2D(64, kernel_size = (3 ,3), activation="relu"),
        layers.MaxPooling2D(pool_size=(2,2)),
        layers.Flatten(),
        layers.Dense(num_classes, activation="softmax")
    ]
 
)
print(modelo.summary())
```

```python
batch_size = 128
epochs = 20
modelo.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
modelo.fit(X_train, y_train, batch_size = batch_size, epochs = epochs, validation_split=0.1)
```

```python
score = modelo.evaluate(X_test, y_test)
print("Erro no teste: ", round(score[0], 5))
print("Acuracia no teste: ", round(score[1], 5))
```

```python
modelo.save('treino_mnist_08062024.keras')
```

```python
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.models import Sequential
```

```python
(train_images, train_labels), (test_images, test_labels) = cifar10.load_data()
```

```python
train_images.shape, train_labels.shape, test_images.shape, test_labels.shape
```

```python
train_images = train_images / 255.0
test_images = test_images / 255.0
```

```python
modelo = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(32,32,3)),
    MaxPooling2D((2,2)),
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D((2,2)),
    Conv2D(64, (3,3), activation='relu'),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(10)
])
```

```python
modelo.compile(optimizer='adam',
                loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                metrics=['accuracy'])
```

```python
def preprocessa_imagem(imagem):
    img = image.load_img(imagem, target_size=(32, 32))
    img_tensor = image.img_to_array(img)
    img_tensor = np.expand_dims(img_tensor, axis=0)
    img_tensor /= 255.0
    return img_tensor
```
